```python
import tkinter as tk
from tkinter import messagebox
import math

class ScientificCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Scientific Calculator")
        self.root.geometry("700x800")
        
        self.expression = ""
        self.text_input = tk.StringVar()
        
        self.create_widgets()
        
    def create_widgets(self):
        # Display
        entry = tk.Entry(self.root, font=('arial', 20, 'bold'), textvariable=self.text_input, bd=30, insertwidth=4, bg="powder blue", justify='right')
        entry.grid(columnspan=4)

        # Buttons
        buttons = [
            '7', '8', '9', 'C',
            '4', '5', '6', '/',
            '1', '2', '3', '*',
            '0', '.', '=', '+',
            '(', ')', 'sin', '-',
            'cos', 'tan', 'sqrt', 'log'
        ]

        row = 1
        col = 0

        for button in buttons:
            if button not in {'=', 'C', 'sin', 'cos', 'tan', 'sqrt', 'log'}:
                tk.Button(self.root, text=button, padx=20, pady=20, bd=8, fg="black", font=('arial', 20, 'bold'), command=lambda btn=button: self.button_click(btn)).grid(row=row, column=col)
            elif button == '=':
                tk.Button(self.root, text=button, padx=20, pady=20, bd=8, fg="black", font=('arial', 20, 'bold'), bg="green", command=self.evaluate).grid(row=row, column=col)
            elif button == 'C':
                tk.Button(self.root, text=button, padx=20, pady=20, bd=8, fg="black", font=('arial', 20, 'bold'), bg="red", command=self.clear).grid(row=row, column=col)
            else:
                tk.Button(self.root, text=button, padx=20, pady=20, bd=8, fg="black", font=('arial', 20, 'bold'), command=lambda func=button: self.function_click(func)).grid(row=row, column=col)
            
            col += 1
            if col > 3:
                col = 0
                row += 1
    
    def button_click(self, button):
        self.expression += str(button)
        self.text_input.set(self.expression)
    
    def function_click(self, func):
        try:
            if func == 'sin':
                self.expression = str(math.sin(math.radians(float(self.expression))))
            elif func == 'cos':
                self.expression = str(math.cos(math.radians(float(self.expression))))
            elif func == 'tan':
                self.expression = str(math.tan(math.radians(float(self.expression))))
            elif func == 'sqrt':
                self.expression = str(math.sqrt(float(self.expression)))
            elif func == 'log':
                self.expression = str(math.log10(float(self.expression)))
            self.text_input.set(self.expression)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear()
    
    def evaluate(self):
        try:
            self.expression = str(eval(self.expression))
            self.text_input.set(self.expression)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear()
    
    def clear(self):
        self.expression = ""
        self.text_input.set("")

if __name__ == "__main__":
    root = tk.Tk()
    calc = ScientificCalculator(root)
    root.mainloop()
```
